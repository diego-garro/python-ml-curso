# Árboles de Regresión

```python
import pandas as pd
```

```python
data = pd.read_csv("../datasets/boston/Boston.csv")
data.head()
```

```python
data.shape
```

```python
colnames = data.columns.values.tolist()
predictors = colnames[:13]
target = colnames[13]
X = data[predictors]
Y = data[target]
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regtree = DecisionTreeRegressor(min_samples_split=30, min_samples_leaf=10, random_state=0)
```

```python
regtree.fit(X, Y)
```

```python
preds = regtree.predict(data[predictors])
```

```python
data["preds"] = preds
```

```python
data[["preds", "medv"]]
```

```python
from sklearn.tree import export_graphviz
with open("resources/boston_rtree.dot", "w") as dotfile:
    export_graphviz(regtree, out_file=dotfile, feature_names=predictors)
    dotfile.close()

import os
from graphviz import Source
file = open("resources/boston_rtree.dot", "r")
text = file.read()
Source(text)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
cv = KFold(n_splits=10, shuffle=True, random_state=1)
scores = cross_val_score(regtree, X, Y, cv=cv, n_jobs=1)
print(scores)
score = np.mean(scores)
print(score)
```
