## Cociente o ratio de probabilidades
Cociente entre los casos de éxito  sobre los de fracaso en el suceso estudiado para cada grupo

```python
display(Math(r'P_m = \ probabilidad\ de\ hacer\ una\ compra\ sabiendo\ que\ es\ un\ hombre'))
display(Math(r'P_m = \ probabilidad\ de\ hacer\ una\ compra\ sabiendo\ que\ es\ una\ mujer'))
display(Math(r'odds\in[0, +\infty]'))
display(Math(r'odds_{purchase, male} = \frac{P_m}{1 - P_m} = \frac{N_{p,m}{\bar p, m}}'))
display(Math(r'odds_{purchase, female} = \frac{P_m}{1 - P_m} = \frac{N_{p,f}{\bar p, f}}'))
```

```python
pm = 121/246
pf = 159/265
odds_m = pm / (1 - pm) # 121/125
odds_f = pf / (1 - pf) # 159/106
```

```python
odds_m
```

```python
odds_f
```

* Si el ratio es superior a 1, es más probable el éxito que el fracaso. Cuanto mayor es el ratio, más probalabilidad de éxito en nuestro suceso.
* Si el ratio es exactamente igual a 1, éxito y fracaso son equiprobables (p=0.5).
* Si el ratio es menor que 1, el fracaso es más probable que el éxito. Cuanto menor es el ratio, menor es la probablidad de éxito del suceso.

```python
display(Math(r'odds_{ratio} = \frac{odds_{purchase, male}}{odds_{purchase, female}}'))
```

```python
odds_ratio = odds_m / odds_f
odds_ratio
```

```python
1 / odds_ratio
```

## La regresión logística desde la regreseión lineal

```python
display(Math(r'y = \alpha + \beta \cdot x'))
display(Math(r'(x, y)\in[-\infty, +\infty]^2'))
```

```python
display(Math(r'y\in\{0, 1\}'))
display(Math(r'p\in [0, 1]'))
display(Math(r'X\in [-\infty, \infty]'))
display(Math(r'P = \alpha + \beta\cdot X'))
```

P es la probabilidad condicionada de éxito o de fracaso condicionada a la presencia de la variable X

```python
display(Math(r'\frac{P}{1 - P} = \alpha + \beta\cdot X\in [0, +\infty]'))
```

```python
display(Math(r'ln\frac{P}{1 - P} = \alpha + \beta\cdot X'))
```

```python
display(Math(r'\begin{cases}\frac{P}{1 - P}\in[0, 1]\Rightarrow ln(\frac{P}{1 - P})\in [-\infty, 0]\\ \frac{P}{1 - P}\in[1, +\infty]\Rightarrow ln(\frac{P}{1 - P})\in [0, +\infty]\end{cases}'))
```

```python
display(Math(r'\begin{equation}ln\frac{P}{1 - P} = \alpha + \beta\cdot X\end{equation}'))
display(Math(r'\begin{equation}\frac{P}{1 - P} = e^(\alpha + \beta\cdot X)\end{equation}'))
display(Math(r'\begin{equation}P = \frac{e^{\alpha + \beta\cdot X}}{1 + e^{\alpha + \beta\cdot X}}\end{equation}'))
display(Math(r'\begin{equation}P = \frac{1}{e^{-(\alpha + \beta\cdot X)}}\end{equation}'))
```

* Si a + bX es muy pequeño (negativo), entonces P tiende a 0
* Si a + bX = 0, P = 0.5
* Si a + bX es muy grande (positivo), entoces P tiende a 1

# Regresión logística múltiple

```python
display(Math(r'\begin{equation}P = \frac{1}{1 + e^{-(\alpha + \sum_{i=1}^n\beta_i\cdot x_i)}}\end{equation}'))
```

```python
display(Math(r'\begin{equation}\vec{\beta} = (\beta_1, \beta_2, \cdots, \beta_n)\end{equation}'))
display(Math(r'\begin{equation}\vec{X} = (x_1, x_2, \cdots, x_n)\end{equation}'))
display(Math(r'\begin{equation}P = \frac{1}{1 + e^{-(\alpha + \vec\beta_i\cdot X)}}\end{equation}'))
```
